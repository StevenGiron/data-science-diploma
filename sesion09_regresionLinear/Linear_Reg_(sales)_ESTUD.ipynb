11. Como se observo en el punto atenrior, la variable Influencer no es numerica (es categorica) por lo que debemos realizarle una transformacion especial para poder utilizarla en nuestro modelo. A esto se le conoce como variables dummies

Crea variables dummy para la columna Influencer. Pista: puedes usar el metodo get_dummies() de pandas. Puedes encontrar mas info [aqui](https://www.sharpsightlabs.com/blog/pandas-get-dummies/)

12. Ahora intenta entrenar de nuevo (esta vez si te deberia funcionar)

```python
X = data2[['TV', 'Radio', 'Social Media', 'Influencer_Macro', 'Influencer_Mega', 'Influencer_Micro', 'Influencer_Nano']]
y = data2['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
lm.fit(X_train, y_train)
```

13. Evalua el modelo con sus distintas metricas (pista, use el metodo predict). Que puedes decir de ellas? 

```python
predictions = lm.predict(X_test)
predictions
```

```python
y_test
```

```python
sns.scatterplot(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

y_test - y Predicted es igual al error, entre más cercano a 0 mas precisión

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

RMSE: se toma Y verdadera y predicted y se eleva al cuadrado, cuando el error es muy grande se castiga el error, viceversa si el error es pequeño. Luego se suma los errores y se saca el promedio, luego se saca la raiz (antes de la raiz es el MSE) para quitarle el cuadrado.   
RMSE es dolares (3.15 millones) un error de 3.15 millones. hacia abajo o hacia arriba.
